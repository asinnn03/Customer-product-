```python
df.hist(bins=10, figsize=(12, 6))
plt.tight_layout()
plt.suptitle("Histogram of Numerical Columns", fontsize=16)
plt.show()
```

```python

cat_cols = df.select_dtypes(include='object').columns
if len(cat_cols) > 0:
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df, x=cat_cols[0])
    plt.title(f"Count Plot of {cat_cols[0]}")
    plt.xticks(rotation=45)
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df.select_dtypes(include='number'))
plt.title("Box Plot of Numerical Columns")
plt.xticks(rotation=45)
plt.show()
```

```python
# Select first categorical column
categorical_columns = df.select_dtypes(include=['object']).columns

if len(categorical_columns) > 0:
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df, x=categorical_columns[0])
    plt.xticks(rotation=45)
    plt.title(f'Count Plot of {categorical_columns[0]}', fontsize=14)
    plt.show()
else:
    print("No categorical columns available for count plot.")
```

```python
if len(cat_cols) > 0:
    df[cat_cols[0]].value_counts().plot.pie(autopct='%1.1f%%', figsize=(6, 6))
    plt.title(f"Pie Chart of {cat_cols[0]}")
    plt.ylabel("")
    plt.show()
```
